```python
import pandas as pd
customers = pd.read_csv("customers.csv")
campaigns = pd.read_csv("campaigns.csv")
transactions = pd.read_csv("transactions.csv")
products = pd.read_csv("products.csv")
customers.head()
campaigns.head()
transactions.head()
products.head()
print("customers:", customers.shape)
print("campaigns:", campaigns.shape)
print("transactions:", transactions.shape)
print("products:", products.shape)
```

```python
print(customers.columns.tolist())
print(campaigns.columns.tolist())
print(transactions.columns.tolist())
print(products.columns.tolist())
print(customers.isnull().sum())
print(campaigns.isnull().sum())
print(transactions.isnull().sum())
print(products.isnull().sum())
```

Convert date columns

```python
customers["signup_date"] = pd.to_datetime(customers["signup_date"], errors="coerce")
campaigns["start_date"] = pd.to_datetime(campaigns["start_date"], errors="coerce")
campaigns["end_date"] = pd.to_datetime(campaigns["end_date"], errors="coerce")
transactions["timestamp"] = pd.to_datetime(transactions["timestamp"], errors="coerce")
products["launch_date"] = pd.to_datetime(products["launch_date"], errors="coerce")
```

```python
transactions_clean = transactions[transactions["refund_flag"] == 0].copy()
transactions_clean.shape
```

```python
merged = transactions_clean.merge(customers, on="customer_id", how="left")
merged = merged.merge(products, on="product_id", how="left")
merged = merged.merge(campaigns, on="campaign_id", how="left")
merged.head()
```
